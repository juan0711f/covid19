# **Introducción**
El estudio de las interacciones entre depredadores y presas ha sido un tema central en la ecología matemática desde principios del siglo XX. Uno de los modelos más influyentes para describir esta dinámica es el modelo de Lotka-Volterra, propuesto de manera independiente por Alfred J. Lotka y Vito Volterra en la década de 1920. Este modelo describe cómo las poblaciones de depredadores y presas interactúan. Sin embargo, el modelo clásico de Lotka-Volterra supone que la población de presas crece de manera exponencial en ausencia de depredadores, lo cual no es realista en la mayoría de los ecosistemas, donde los recursos son limitados.

Para abordar esta limitación, se introdujo el modelo de depredador-presa logístico, que incorpora un crecimiento logístico para la población de presas. Este modelo tiene en cuenta que los recursos disponibles para las presas son finitos, lo que impone un límite superior $K$ (conocido como capacidad de carga) en su crecimiento poblacional. Esta modificación hace que el modelo sea más realista y adecuado para describir sistemas ecológicos donde la disponibilidad de recursos es un factor crítico. Para este modelo, los depredadores se denota por $P$ la cantidad de depredadores y por $N$ la cantidad de presas.

Este modelo está dado por las siguientes ecuaciones:

$$
\begin{cases}
\frac{dN}{dt} = rN\left( 1 - \frac{N}{K} \right) - \alpha NP = f(N,P)
\\ \\
\frac{dP}{dt} = \beta \alpha NP - \delta P = g(N,P)
\end{cases}  
$$

donde $r$ es la tasa de crecimiento de las presas, $K$ es la capacidad de carga la cual representa el número máximo de presas que el ambiente puede sostener, $\alpha$ es el coeficiente que mide la tasa de depredación, $\beta$ es la eficiencia con la que los depredadores convierten las presas consumidas en nuevos individuos y $\delta$ es la tasa de mortalidad de los depredadores.

Observe en el modelo, el término $rN\left(1 - \frac{N}{K}\right)$ representa el crecimiento logístico. A medida que $N$ se acerca a $K$, el crecimiento se ralentiza, reflejando la limitación de recursos. Asi mismo, para la ecuación de presas, el término $\alpha NP$ indica que la tasa a la que las presas son capturadas es proporcional tanto a la cantidad de presas como a la de depredadores y para el la ecuación de depredadores, el término $\beta \alpha NP$ indica que los depredadores muestra que su crecimiento depende de la disponibilidad de presas y que, en ausencia de presas, su población decrece a una tasa $\delta$.


# **Análisis de estabilidad**
El análisis de estabilidad es una herramienta fundamental para comprender el comportamiento a largo plazo del sistema. Este análisis se centra en identificar y estudiar los puntos de equilibrio, es decir, los valores de $N$ y $P$ para los cuales las derivadas temporales se anulan:

$$ \frac{dP}{dt} = 0 \quad \text{y} \quad \frac{dN}{dt} = 0 $$

es decir,

\begin{equation}\tag{1}
rN\left( 1 - \frac{N}{K} \right) - \alpha NP = 0,
\end{equation}
\begin{equation}
\beta \alpha NP - \delta P = 0.
\tag{2}
\end{equation}

De $(2)$, tenemos que

\begin{equation}
\begin{split}
P(\beta \alpha N - \delta) = 0,
\end{split}
\end{equation}

lo cual nos lleva a tener

\begin{equation}\tag{3}
P = 0,
\end{equation}
o en segundo caso,
\begin{equation}\tag{4}
N = \frac{\delta}{\beta \alpha}.
\end{equation}

**Caso 1:** $P=0$

Si $P=0$, entonces la ecuación $(1)$ quedaría

$$rN\left( 1 - \frac{N}{K} \right)  = 0$$

Lo cual implica que $N=0$ o $N = K$. Así, obtendríamos los siguientes puntos críticos:


1.   $(N^*,P^*) = (0,0)$,  el cual representa la extinción de ambas especies.
2.   $(N^*,P^*) = (K,0)$,  el cual representa la extinción del depredador y la presa alcanzando su capacidad de carga $K$.

**Caso 2:** $N = \frac{\delta}{\beta \alpha}.$

Sustituyendo $N = \frac{\delta}{\beta \alpha}$ en $(1)$ obtenemos

$$ r\left( \frac{\delta}{\beta \alpha} \right)\left( 1 - \frac{\left( \frac{\delta}{\beta \alpha} \right)}{K} \right) - \alpha \left( \frac{\delta}{\beta \alpha} \right)P = 0 $$

despejando $P$ se tiene que

$$ P = \frac{r}{\alpha}\left( 1 - \frac{\delta}{\beta \alpha K} \right) $$

por lo tanto, se obtendría el punto crítico

3. $(N^*,P^*) = \left( \frac{\delta}{\beta \alpha}, \frac{r}{\alpha}\left( 1 - \frac{\delta}{\beta \alpha K} \right) \right)$, el cual representa un equilibrio donde ambas especies coexisten


Ahora, para analizar la estabilidad de los puntos críticos, calculamos la matriz Jacobiana, dada por
$$ J(N,P) = \left( \begin{array}{cc} \frac{df}{dN} & \frac{df}{dP} \\ \frac{dg}{dN} & \frac{dg}{dP} \end{array} \right) = \left( \begin{array}{cc} r\left(1-\frac{2N}{K}\right)-\alpha P & -\alpha N \\ \beta \alpha P & \beta \alpha N - \delta \end{array} \right) $$

1. Para $(N^*, P^*) = (0,0)$, se tiene que
$$ J(0,0) = \left(\begin{array}{cc}r & 0 \\ 0 & -\delta \end{array}\right)$$

por lo tanto los valores propios son $\lambda_1 = r$ y $\lambda_2 = -\delta$. Por lo tanto, el punto es un punto silla o punto inestable.

2. Para $(N^*, P^*) = (K, 0)$, se tiene que
$$ J(K,0) = \left( \begin{array}{cc}-r & -\alpha K \\ 0 & \beta \alpha K - \delta\end{array} \right)$$

Lo cual, daría el valor propio $\lambda_1 = -r$ y $\lambda_2 = \beta \alpha K - \delta$.

Si $\beta \alpha K < \delta$, entonces $\lambda_2 < 0$, y $(K,0)$ sería un punto estable. Se podría interpretar esta caso donde el depredador se extinguirá ($\delta$ es muy grande) y la presa alcanzará su capacidad de carga $K$.

Si $\beta \alpha K > \delta$, entonces $\lambda_2 > 0$ y $(K,0)$ sería un punto inestable, lo que sugiere que el sistema no permanecerá en este estado y podría evolucionar hacia el otro punto crítico.

3. Para $(N^*,P^*) = \left( \frac{\delta}{\beta \alpha}, \frac{r}{\alpha}\left( 1 - \frac{\delta}{\beta \alpha K} \right) \right)$

Observe que si $\frac{\delta}{\beta \alpha K} > 1$, entonces $P < 0$, lo cual no es posible ya que $P$ representa una cantidad no negativa. Por lo tanto, el punto existe solo si $\frac{\delta}{\beta \alpha K} < 1$.

Así, se tiene que

$$J(N^*, P^*) = \left( \begin{array}{cc} r\left(1-\frac{2N^*}{K}\right)-\alpha P^* & - \alpha N^* \\ \beta \alpha P^* & \beta \alpha N^*-\delta \end{array} \right)$$

Sustituyendo $N^* = \frac{\delta}{\beta \alpha}$, $ P^* = \frac{r}{\alpha}\left( 1 - \frac{\delta}{\beta \alpha K} \right)$ y simplificando, se obtiene

$$J(N^*, P^*) = \left(\begin{array}{cc}-\frac{r\delta}{\beta \alpha K} & -\frac{\delta}{\beta} \\ r\beta \left(1-\frac{\delta}{\beta \alpha K}\right) & 0\end{array}\right).$$

Con el objetivo de encontrar los valores propios, tenemos la ecuación característica

$$\lambda^2 + \left( \frac{r \delta}{\beta \alpha K} \right)\lambda + r \delta \left(1 - \frac{\delta}{\beta \alpha K}\right) = 0$$

Por la gran dependencia de los parámetros, no podemos determinar el tipo específico de valor crítico.

# **Discretización del modelo**

En esta sección, con el fin de analizar el comportamiento, se procederá a realizar una implementación numérica del modelo. Para esto, se usará el método numérico de diferencias finitas, el cual consiste en una aproximación de las derivadas que aparecen en un problema de ecuaciones diferenciales ordinarias, reduciéndolo a resolver un sistema lineal en un número limitado de puntos seleccionado. Por lo tanto, presentamos una discretización del modelo depredador-presa logístico para ser utilizado futuras simulaciones.
Para esto, se asume una partición de $[0,T]$ con paso de tiempo $k = T/N : (t_n = nk)_{n=0}^{n=N}$. Así, se considera el siguiente esquema numérico de primer orden en el tiempo desacoplado.

**Inicialización:** Sea $[N_0, P_0]$

**Paso de tiempo n:** Dado el vector $[N_{n-1}, P_{n-1}]$, calcular $[N_{n}, P_n]$ tal que

$\left \{ \begin{array}{l}\frac{N_n - N_{n-1}}{k}=r N_{n-1}\left( 1 - \frac{N_{n-1}}{K}\right)-\alpha N_{n-1} P_{n-1}\\\\ \frac{P_n - P_{n-1}}{k}= P_{n-1} + \beta \alpha N_{n-1}P_{n-1} - \delta P_{n-1} \end{array}\right .$

# **Simulaciones**
A continuación, se mostrará el código desarrollado en python implementando el esquema numérico y posteriormente se mostrarán distintas simulaciones para así observar el comportamiento del modelo discreto.

Inicialmente, se hará una simulación del modelo para distintos parámetros. Así, se podrá observar algunos comportamientos del esquema.

Para estas simulaciones, se asumirá una población inicial de depredadores de 50 y de presas 80.

El objetivo de las siguientes simulaciones es observar el comportamiento de los puntos críticos y del modelo para distintos valores de los parámetros. En las dos últimas gráficas, se escogieron parámetros específicos que hagan tanto positivo como negativo el discriminante:


$$ \Delta = \left( \frac{r \delta}{\beta \alpha K} \right)^2 - 4\left(1 - \frac{\delta}{\beta \alpha K}\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Se usa el estilo "ggplot"
plt.style.use('ggplot')

# Definición de los 6 conjuntos de parámetros
all_param_sets = [
    {"r": 0.5, "K": 100, "alpha": 0.01, "beta": 0.1,  "delta": 0.1,  "desc": "Caso 1"},
    {"r": 1.0, "K": 100, "alpha": 0.015, "beta": 0.1, "delta": 0.15, "desc": "Caso 2"},
    {"r": 1.5, "K": 120, "alpha": 0.02, "beta": 0.15, "delta": 0.1,  "desc": "Caso 3"},
    {"r": 2.0, "K": 150, "alpha": 0.03, "beta": 0.1, "delta": 0.2,  "desc": "Caso 4"},
    {"r": 1.0, "K": 80,  "alpha": 0.02, "beta": 0.1, "delta": 0.1,  "desc": "Caso 5 (Δ > 0)"},
    {"r": 1.0, "K": 500, "alpha": 0.02, "beta": 0.1, "delta": 0.1,  "desc": "Caso 6 (Δ < 0)"}
]

# Parámetros de la simulación temporal
T = 150       # Tiempo total
dt = 0.1      # Paso temporal
n_steps = int(T/dt)
t = np.linspace(0, T, n_steps)

# Función para simular el sistema usando el método de Euler
def modelo(r, K, alpha, beta, delta):
    N = np.zeros(n_steps)  # Población de presas
    P = np.zeros(n_steps)  # Población de depredadores
    # Condiciones iniciales (consistentes para todos los casos)
    N[0] = 50
    P[0] = 80
    for i in range(n_steps - 1):
        dN = r * N[i] * (1 - N[i] / K) - alpha * N[i] * P[i]
        dP = beta * alpha * N[i] * P[i] - delta * P[i]
        N[i+1] = N[i] + dt * dN
        P[i+1] = P[i] + dt * dP
    return N, P

# Función para obtener datos del campo vectorial
def campo_vectorial(r, K, alpha, beta, delta, N_range, P_range, n_points=50):
    N_vals = np.linspace(N_range[0], N_range[1], n_points)
    P_vals = np.linspace(P_range[0], P_range[1], n_points)
    N_grid, P_grid = np.meshgrid(N_vals, P_vals)
    dNdt = r * N_grid * (1 - N_grid / K) - alpha * N_grid * P_grid
    dPdt = beta * alpha * N_grid * P_grid - delta * P_grid
    return N_grid, P_grid, dNdt, dPdt

# Crear una figura con 6 filas y 2 columnas (cada fila: evolución temporal | campo vectorial)
fig, axs = plt.subplots(len(all_param_sets), 2, figsize=(14, 36))

# Iterar sobre cada conjunto de parámetros
for i, params in enumerate(all_param_sets):
    r     = params["r"]
    K     = params["K"]
    alpha = params["alpha"]
    beta  = params["beta"]
    delta = params["delta"]
    desc  = params["desc"]

    # Simulación temporal
    N, P = modelo(r, K, alpha, beta, delta)

    # --- Columna izquierda: Evolución temporal ---
    axs[i, 0].plot(t, N, label="Presas", color="blue")
    axs[i, 0].plot(t, P, label="Depredadores", color="red")
    axs[i, 0].set_title(f"{desc} - Evolución temporal\nr = {r}, K = {K}, α = {alpha}, β = {beta}, δ = {delta}")
    axs[i, 0].set_xlabel("Tiempo")
    axs[i, 0].set_ylabel("Población")
    axs[i, 0].legend()
    axs[i, 0].grid(True)

    # --- Columna derecha: Campo vectorial ---
    # Definir rangos para el campo vectorial:
    # N: de 0 a K+50, P: de 0 a (valor máximo de P en la simulación + 50)
    N_range_plot = (0, K + 50)
    P_range_plot = (0, np.max(P) + 50)
    N_grid, P_grid, dNdt, dPdt = campo_vectorial(r, K, alpha, beta, delta, N_range_plot, P_range_plot, n_points=50)
    strm = axs[i, 1].streamplot(N_grid, P_grid, dNdt, dPdt, density=1.2, color='black', linewidth=1, arrowsize=1.5)

    # Marcar puntos críticos:
    # 1. Extinción: (0,0)
    axs[i, 1].scatter(0, 0, color='magenta', s=100, label="(0,0)")
    # 2. Solo presas: (K,0)
    axs[i, 1].scatter(K, 0, color='cyan', s=100, label="(K,0)")
    # 3. Coexistencia: se calcula como N* = δ/(β·α) y P* = r/α (1 - N*/K), siempre que N* < K.
    N_star = delta / (beta * alpha)
    if N_star < K:
        P_star = (r / alpha) * (1 - (N_star / K))
        axs[i, 1].scatter(N_star, P_star, color='orange', s=100, label="(N*,P*)")
        axs[i, 1].annotate("(N*,P*)", (N_star, P_star), textcoords="offset points", xytext=(5,5), color='orange', fontsize=12)
    axs[i, 1].annotate("(0,0)", (0,0), textcoords="offset points", xytext=(5,-10), color='magenta', fontsize=12)
    axs[i, 1].annotate("(K,0)", (K,0), textcoords="offset points", xytext=(5,5), color='cyan', fontsize=12)

    axs[i, 1].set_title("Campo vectorial")
    axs[i, 1].set_xlabel("Presas (N)")
    axs[i, 1].set_ylabel("Depredadores (P)")
    axs[i, 1].set_xlim(N_range_plot)
    axs[i, 1].set_ylim(P_range_plot)
    axs[i, 1].legend(loc='upper right')
    axs[i, 1].grid(True)

plt.tight_layout()
plt.show()
```
